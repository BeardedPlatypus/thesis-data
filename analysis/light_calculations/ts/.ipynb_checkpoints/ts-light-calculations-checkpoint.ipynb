```python
import json
from pathlib import Path

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf', 'svg')
matplotlib.rcParams['svg.fonttype'] = 'none'
```

```python
RAW_DATA_ROOT = Path(r"C:\Users\Monthy\Documents\projects\thesis\thesis-data-suite\data\raw\lc-summed")
ANALYSIS_ROOT = Path(r"C:\Users\Monthy\Documents\projects\thesis\thesis-data-suite\data\analysis\lc-summed\ts")
```

```python
scenes = [("pipers-alley",  "1044#3x_3y_2z"),
          ("ziggurat-city", "1170#3x_3y_2z"),
          ("spaceship-indoor", "1260#3x_3y_2z"),
         ]

tile_sizes = [8, 16, 32, 64]
resolutions = [320, 640, 960, 1280, 1920, 2560]
pipelines = ["DEFERRED"]

res_path = RAW_DATA_ROOT / Path("resolution/ts")
```

```python
def json_to_dataframe(json_dic : dict) -> pd.DataFrame:
    '''
    Convert a json data file spit out by nTiled into something just slightly more useful
    '''
    # get all functions tracked from first frame
    frames = json_dic["frames"]
    
    frames_i = {}
    frames_n_calc = {}
    for f in frames:
        f_i = int(f["frame"])
        
        frames_i[f_i] = f_i
        frames_n_calc[f_i] = int(f["n_calc"])
        
    data_series = { "index": pd.Series(frames_i, name="index"),
                    "n_calc": pd.Series(frames_n_calc, name="n_calc")
                  }
    
    dataframe = pd.DataFrame(data_series)
    return dataframe
```

```python
def plot_resolution(scene: str, pipeline_type: str):
    # Build empty dataframes
    # -----------------------------------------------------------------------------------
    attenuated_data_folder = RAW_DATA_ROOT / Path("resolution/ns/{}/{}".format(pipeline_type, scene))
    
    df_attenuated = pd.DataFrame([], columns=["n_calc_per_frame", "resolution"])
    index = 0
    
    for res in resolutions:
        attenuated_file_path = attenuated_data_folder / Path("{}/lc_data.json".format(res))
        with open(str(attenuated_file_path), 'r') as f:
            data_json = json.loads(f.read())
        df_att = json_to_dataframe(data_json)
        df_val = df_att["n_calc"].mean() / (1 * 10 ** 5)
        df_attenuated.loc[index] = [df_val, res]
        index += 1
    
    df_tiles = {}
    for x in tile_sizes:
        df_tiles[x] = pd.DataFrame([], columns=["n_calc_per_frame", "resolution"])

    # Acquire data
    # -----------------------------------------------------------------------------------
    tiled_data_folder = res_path / Path("{}/{}".format(pipeline_type, 
                                                          scene))
    
    index = 0
    
    for x in tile_sizes:
        for res in resolutions:
            tiled_file_path = tiled_data_folder / Path("{}/{}/lc_data.json".format(x, res))
            with open(str(tiled_file_path), 'r') as f:
                data_json = json.loads(f.read())
            df_tiled = json_to_dataframe(data_json)
                
            df_val = df_tiled["n_calc"].mean() / (1 * 10 ** 6) #/ (res * res)               
                
            df_tiles[x].loc[index] = [df_val, res]            
            index += 1

    # Plot resolution
    # -----------------------------------------------------------------------------------
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    sns.regplot("resolution", "n_calc_per_frame",
                data=df_attenuated,
                ax=ax,
                #fit_reg=False,
                label= "naief",
                order=2)
    
    for x in tile_sizes:
        sns.regplot("resolution", "n_calc_per_frame", 
                    data=df_tiles[x], 
                    #fit_reg=False,
                    ax=ax,
                    label="Tiled - {}".format(x),
                    order=2)
    
    ax.set(xlabel='Resolutie (px)', ylabel='Aantal lichtberekeningen (10^6)')
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)
    
    ax.set(ylim=(-49, None))
    
    plt.show()
```

```python
def export_resolution(scene: str, pipeline_type: str):
    # Build empty dataframes
    # -----------------------------------------------------------------------------------
    attenuated_data_folder = RAW_DATA_ROOT / Path("resolution/ns/{}/{}".format(pipeline_type, scene))
    
    df_attenuated = pd.DataFrame([], columns=["n_calc_per_frame", "resolution"])
    index = 0
    
    for res in resolutions:
        attenuated_file_path = attenuated_data_folder / Path("{}/lc_data.json".format(res))
        with open(str(attenuated_file_path), 'r') as f:
            data_json = json.loads(f.read())
        df_att = json_to_dataframe(data_json)
        df_val = df_att["n_calc"].mean() / (1 * 10 ** 5)
        df_attenuated.loc[index] = [df_val, res]
        index += 1
    
    df_tiles = {}
    for x in tile_sizes:
        df_tiles[x] = pd.DataFrame([], columns=["n_calc_per_frame", "resolution"])

    # Acquire data
    # -----------------------------------------------------------------------------------
    tiled_data_folder = res_path / Path("{}/{}".format(pipeline_type, 
                                                          scene))
    
    index = 0
    
    for x in tile_sizes:
        for res in resolutions:
            tiled_file_path = tiled_data_folder / Path("{}/{}/lc_data.json".format(x, res))
            with open(str(tiled_file_path), 'r') as f:
                data_json = json.loads(f.read())
            df_tiled = json_to_dataframe(data_json)
                
            df_val = df_tiled["n_calc"].mean() / (1 * 10 ** 6) #/ (res * res)               
                
            df_tiles[x].loc[index] = [df_val, res]            
            index += 1

    # Plot resolution
    # -----------------------------------------------------------------------------------
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    sns.regplot("resolution", "n_calc_per_frame",
                data=df_attenuated,
                ax=ax,
                #fit_reg=False,
                label= "naief",
                order=2)
    
    for x in tile_sizes:
        sns.regplot("resolution", "n_calc_per_frame", 
                    data=df_tiles[x], 
                    #fit_reg=False,
                    ax=ax,
                    label="Tiled - {}".format(x),
                    order=2)
    
    ax.set(xlabel='Resolutie (px)', ylabel='Aantal lichtberekeningen (10^6)')
    
    ax.set(ylim=(-49, None))
    
    ax.xaxis.labelpad = 15
    ax.yaxis.labelpad = 25

    # Export file
    # -------------------------------------------------------------
    path = ANALYSIS_ROOT / Path("graphs/resolution/{}".format(pipeline_type))
    
    if not path.exists() or path.is_file():
        path.mkdir(parents=True)      
        
    output_path = path / Path("resolution_{}.svg".format(scene))
    fig.savefig(str(output_path), bbox_inches='tight')
    plt.close(fig)
```

```python
scenelets = ["pipers-alley",
             "ziggurat-city",
             "spaceship-indoor",
            ]

pts = ["deferred",]
for pt in pts:
    for scene in scenelets:
        plot_resolution(scene, pt)
        export_resolution(scene, pt)
```
