```python
def json_to_dataframe(json_dic : dict) -> pd.DataFrame:
    '''
    Convert a json data file spit out by nTiled into something just slightly more useful
    '''
    # get all functions tracked from first frame
    frames = json_dic["frames"]
    functions = list(x for x in frames[0]["functions"])
    
    # build empty dictionary which will be loaded with data from frames
    data_raw = { fn : {} for fn in functions }    
    frames_i = {}
    
    for f in frames:
        f_i = int(f["frame"])
        frames_i[f_i] = f_i
        for fn in functions:
            if fn in f["functions"]:
                data_raw[fn][f_i] = float(f["functions"][fn])
            
    data_series = {fn : pd.Series(data_raw[fn], name=fn) for fn in functions }
    data_series["index"] = pd.Series(frames_i, name="index")
    
    dataframe = pd.DataFrame(data_series)
    return dataframe
```

```python
def average_tests(df_l : list) -> pd.DataFrame:
    frame_mean = pd.concat(df_l)
    by_row_index = frame_mean.groupby(frame_mean.index)
    df_means = by_row_index.mean()
    
    return df_means


def export_frames_plot(pipeline_type: str, scene: str, l_def: str, res: int, tile_size : int):
    # Obtain data
    # ---------------------------------------------------------------------------------------------------
    tiled_data_folder = RAW_DATA_ROOT_TILED / Path("{}/{}/{}/{}/{}".format(pipeline_type, 
                                                                            scene, 
                                                                            l_def, 
                                                                            res, 
                                                                            tile_size))
    attenuated_data_folder = RAW_DATA_ROOT_ATTENUATED / Path("{}/{}/{}/{}".format(pipeline_type,
                                                                                 scene,
                                                                                 l_def,
                                                                                 res))
    
    attenuated_dfs = []
    tiled_dfs = []
    
    for i in range(N_TESTS):
        tiled_file_path = tiled_data_folder / Path("data_{}.json".format(i))
        with open(str(tiled_file_path), 'r') as f:
            data_json = json.loads(f.read())        
        tiled_dfs.append(json_to_dataframe(data_json))
        
        attenuated_file_path = attenuated_data_folder / Path("data_{}.json".format(i))
        with open(str(attenuated_file_path), 'r') as f:
            data_json = json.loads(f.read())
        attenuated_dfs.append(json_to_dataframe(data_json))  
    
    # Plot Data
    # ---------------------------------------------------------------------------------------------------
    df_attenuated = average_tests(attenuated_dfs)
    df_tiled = average_tests(tiled_dfs)
    
    if pipeline_type == "forward":
        render_func_tiled = ["ForwardTiledShader::renderObjects"]
        render_func_att = ["ForwardAttenuatedShader::renderObjects"]
    elif pipeline_type == "deferred":
        render_func_tiled = ["DeferredTiledShader::renderGeometryPassObjects",
                             "DeferredTiledShader::renderLightPassObjects", 
                      ]
        render_func_att = ["DeferredAttenuatedShader::renderGeometryPassObjects",
                           "DeferredAttenuatedShader::renderLightPassObjects", 
                          ]
    else:
        render_func_tiled = []
        render_func_att = []

    
    col_list = list(x for x in df_tiled if (x != 'index') and (not x in render_func_tiled))
    
    total = render_func_tiled + ["grid"]
    df_tiled.loc[:, 'grid'] = df_tiled[col_list].sum(axis=1)
    df_tiled.loc[:, 'total'] = df_tiled[total].sum(axis=1)
    
    df_attenuated.loc[:, 'total'] = df_attenuated[render_func_att].sum(axis=1)
    
        
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    sns.regplot("index", "total", 
                data=df_attenuated, fit_reg=False, 
                ax=ax,
                label="Attenuated")
    sns.regplot("index", "grid", 
                data=df_tiled, fit_reg=False, 
                ax=ax,
                label="Tiled - Rooster opbouw")
    sns.regplot("index", "total", 
                data=df_tiled, fit_reg=False, 
                ax=ax,
                label="Tiled - Totaal")
    
    ax.set(xlabel='Frames', ylabel='Executietijd (ms)')
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)


    # Export file
    # -------------------------------------------------------------
    path = ANALYSIS_ROOT / Path("graphs/frames/{}".format(pipeline_type))
    
    if not path.exists() or path.is_file():
        path.mkdir(parents=True)
        
    output_path = path / Path("frames_{}_{}_{}_{}.svg".format(scene, 
                                                              l_def.split("#")[0],                                                                  
                                                              res, 
                                                              tile_size))
    fig.savefig(str(output_path), bbox_inches='tight')
    plt.close(fig)
```

```python
def export_resolution_plot(scene: str, l_def: str, pipeline_type: str):
    # Build empty dataframes
    # -----------------------------------------------------------------------------------
    df_attenuated = pd.DataFrame([], columns=["execution_time", "resolution"])
    df_tiles = {}
    for x in TILE_SIZES:
        df_tiles[x] = pd.DataFrame([], columns=["execution_time", "resolution"])

    # Acquire data
    # -----------------------------------------------------------------------------------
    tiled_data_folder = RAW_DATA_ROOT_TILED / Path("{}/{}/{}".format(pipeline_type, 
                                                                           scene, 
                                                                           l_def))
    attenuated_data_folder = RAW_DATA_ROOT_ATTENUATED / Path("{}/{}/{}".format(pipeline_type,
                                                                                 scene,
                                                                                 l_def))

    index = 0
    
    for res in RESOLUTIONS_TILED:
        for i in range(3):
            attenuated_file_path = attenuated_data_folder / Path("{}/data_{}.json".format(res, i))
            with open(str(attenuated_file_path), 'r') as f:
                data_json = json.loads(f.read())        
            df_att = json_to_dataframe(data_json)
            
            col_list = list(x for x in df_att if (x != 'index'))
            df_val = df_att[col_list].sum(axis=1).mean()
            
            df_attenuated.loc[index] = [df_val, res]
                        
            for x in TILE_SIZES:
                tiled_file_path = tiled_data_folder / Path("{}/{}/data_{}.json".format(res, x, i))
                with open(str(tiled_file_path), 'r') as f:
                    data_json = json.loads(f.read())
                df_tiled = json_to_dataframe(data_json)
                
                col_list = list(x for x in df_tiled if (x != 'index'))
                df_val = df_tiled[col_list].sum(axis=1).mean()                
                
                df_tiles[x].loc[index] = [df_val, res]            
            index += 1

    # Plot resolution
    # -----------------------------------------------------------------------------------
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    sns.regplot("resolution", "execution_time", 
                data=df_attenuated, #fit_reg=False,
                ax=ax,
                label="Attenuated",
                x_estimator=np.mean,
                order=2,
                ci=100)
    for x in TILE_SIZES:
        sns.regplot("resolution", "execution_time", 
                    data=df_tiles[x], #fit_reg=False,
                    ax=ax,
                    label="Tiled - {}".format(x),
                    x_estimator=np.mean,
                    order=2,
                    ci=100)
    
    ax.set(xlabel='Resolutie (px)', ylabel='Executietijd (ms)')
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)
    
    ax.set(ylim=(-49, None))
    
    # Export file
    # -------------------------------------------------------------
    path = ANALYSIS_ROOT / Path("graphs/resolution/{}".format(pipeline_type))
    
    if not path.exists() or path.is_file():
        path.mkdir(parents=True)
        
    output_path = path / Path("resolution_{}_{}.svg".format(scene, 
                                                            l_def.split("#")[0]))
    fig.savefig(str(output_path), bbox_inches='tight')
    plt.close(fig)
```

```python
scenelets = {"pipers-alley" : "1044#3x_3y_2z",
             "ziggurat-city" : "1170#3x_3y_2z",
             "spaceship-indoor" :"1260#3x_3y_2z",
            }

pts = ["forward", "deferred"]
for pt in pts:
    for scene in scenelets:
        export_resolution_plot(scene, scenelets[scene], pt)
```

```python
def export_lights_plot(scene: str, res: int, pipeline_type: str):
    # Build empty dataframes
    # -----------------------------------------------------------------------------------
    df_attenuated = pd.DataFrame([], columns=["execution_time", "n_lights"])
    df_tiles = {}
    for x in TILE_SIZES:
        df_tiles[x] = pd.DataFrame([], columns=["execution_time", "n_lights"])

    # Acquire data
    # -----------------------------------------------------------------------------------
    tiled_data_folder = RAW_DATA_ROOT_TILED / Path("{}/{}".format(pipeline_type, 
                                                                     scene))
    attenuated_data_folder = RAW_DATA_ROOT_ATTENUATED / Path("{}/{}".format(pipeline_type,
                                                                            scene))

    index = 0
    
    for ldef in SCENES_LIGHTS[scene]:
        n_lights = int(ldef.split("#")[0])
        for i in range(3):
            attenuated_file_path = attenuated_data_folder / Path("{}/{}/data_{}.json".format(ldef, res, i))
            with open(str(attenuated_file_path), 'r') as f:
                data_json = json.loads(f.read())        
            df_att = json_to_dataframe(data_json)
            
            col_list = list(x for x in df_att if (x != 'index'))
            df_val = df_att[col_list].sum(axis=1).mean()
            
            df_attenuated.loc[index] = [df_val, n_lights]
                        
            for x in TILE_SIZES:
                tiled_file_path = tiled_data_folder / Path("{}/{}/{}/data_{}.json".format(ldef, res, x, i))
                with open(str(tiled_file_path), 'r') as f:
                    data_json = json.loads(f.read())
                df_tiled = json_to_dataframe(data_json)
                
                col_list = list(x for x in df_tiled if (x != 'index'))
                df_val = df_tiled[col_list].sum(axis=1).mean()                
                
                df_tiles[x].loc[index] = [df_val, n_lights]            
            index += 1

    # Plot resolution
    # -----------------------------------------------------------------------------------
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    sns.regplot("n_lights", "execution_time", 
                data=df_attenuated, #fit_reg=False,
                ax=ax,
                label="Attenuated",
                x_estimator=np.mean,
                order=1,
                ci=100)
    for x in TILE_SIZES:
        sns.regplot("n_lights", "execution_time", 
                    data=df_tiles[x], #fit_reg=False,
                    ax=ax,
                    label="Tiled - {}".format(x),
                    x_estimator=np.mean,
                    order=1,
                    ci=100)
    
    ax.set(xlabel='Aantal lichten', ylabel='Executietijd (ms)')
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)
    
    # Export file
    # -------------------------------------------------------------
    path = ANALYSIS_ROOT / Path("graphs/lights/{}".format(pipeline_type))
    
    if not path.exists() or path.is_file():
        path.mkdir(parents=True)
        
    output_path = path / Path("lights_{}_{}.svg".format(scene, res))
    fig.savefig(str(output_path), bbox_inches='tight')
    plt.close(fig)
```
